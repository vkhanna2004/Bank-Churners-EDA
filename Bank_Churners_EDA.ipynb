### 9.	Plot a percentage segment bar graph between Education_Level and Attrition_Flag of the customers.(10 point)

```python
percentage_data = bank.groupby(['Education_Level', 'Attrition_Flag']).size().unstack().fillna(0)
total = percentage_data.sum(axis=1)
percentage_data['Attrition_Percentage'] = (percentage_data['Attrited Customer'] / total) * 100
percentage_data['Existing_Percentage'] = (percentage_data['Existing Customer'] / total) * 100

plt.figure(figsize=(10, 6))
index = percentage_data.index

plt.bar(index, percentage_data['Attrition_Percentage'], label='Attrited Customer',color='red',  alpha=0.7)
plt.bar(index, percentage_data['Existing_Percentage'], label='Existing Customer', color='blue', alpha=0.7,bottom=percentage_data['Attrition_Percentage'])

plt.xlabel('Education Level')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
```

### 10.	Plot a percentage segment bar graph between Income_Category and Attrition_Flag of the customers.(10 point)

```python
percentage_data = bank.groupby(['Income_Category', 'Attrition_Flag']).size().unstack().fillna(0)
total=percentage_data.sum(axis=1)

percentage_data['Attrition_Percentage']=percentage_data['Attrited Customer']/total*100
percentage_data['Existing_Percentage']=percentage_data['Existing Customer']/total*100

plt.bar(percentage_data.index,percentage_data['Attrition_Percentage'],color='red', label='Attrited Customer',alpha=0.6)
plt.bar(percentage_data.index, percentage_data['Existing_Percentage'], color='blue',label='Existing Customer',alpha=0.6,bottom=percentage_data['Attrition_Percentage'])

plt.xlabel('Income_Category')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend()
```
